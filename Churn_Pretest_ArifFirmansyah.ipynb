## Q3. Ada berapa kah jumlah pelanggan yg stay dan churn? dan jumlah rationya?

```
stayed = churn.groupby('class').phone_number.count()[0]
churned = churn.groupby('class').phone_number.count()[1]
churn_ratio = churned/(stayed+churned)*100

print('Jumlah pelanggan yg stay :', stayed)
print('Jumlah pelanggan yg churn :', churned)
print('Ratio churn :', churn_ratio, "%")
```
